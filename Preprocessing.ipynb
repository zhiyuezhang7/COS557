```python
import anndata as ad
```

```python
adata = sc.read_h5ad(dir + 'Integrated_normalized.h5ad')
adata
```

```python
adata_t = ad.AnnData(adata.X.T)
adata_t
```

```python
adata_t.X
```

```python
# add metadata (cell information)
mdata = pd.read_csv(dir + 'metadata.csv', index_col=0)
mdata
```

```python
adata_t.obs = mdata
adata_t
```

```python
# add gene information
genes_f = open(dir + "genes.txt", "r") 
genes = genes_f.read() 
gene_list = genes.split("\n") 
print(gene_list) 
genes_f.close() 
```

```python
gene_list = gene_list[0:len(gene_list)-1]
len(gene_list)
```

```python
gene_list1 = [gene.replace('"', '').replace('[1] ', '') for gene in gene_list]
gene_list1
```

```python
adata_t.var['gene_name'] = gene_list1
adata_t
```

```python
adata_t.write_h5ad(dir + 'Integrated_normalized_w_metadata.h5ad')
```

```python
# lastly, make label file for learning; here we have two levels of cell type, one more specific than the other
labels = adata_t.obs[['Cell_Cluster_level1','Cell_Cluster_level2']]
labels
```

```python
labels.to_csv(dir + 'cell_type_labels.csv')
```
